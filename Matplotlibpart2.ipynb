```python
# Histogram
# Subplot
# Scatter plot

import matplotlib.pyplot as plt
import numpy as np
```

```python
a = np.array([23,45,67,90])
b = np.array([78,-90,34,222])
plt.scatter(a,b,c = 'red')
plt.xlabel('marks')
plt.ylabel('subject')
```

```python
#histogram
x = [34,56,7,890,35,23]
plt.hist(x)
```

```python
h = np.array([87,5,43,56,73])
plt.hist(h,bins = [0,25,50,75,100])
```

```python
h = np.array([87,5,43,56,73])
plt.hist(h,bins = 10)
```

```python
h = np.array([87,5,43,56,73])
plt.hist(h,bins = 10,rwidth=0.8)
plt.xlabel('abc')
```

```python
h = np.array([87,5,43,56,73])
plt.hist(h,bins = 2)
```

```python
# subplot

a = [45,67,56]
b = [34,67,90]
plt.subplot(2,1,1)
plt.plot(a,b)


x = [23,56,89,34]
y = [45,67,0,34]
plt.subplot(2,1,2)
plt.plot(x,y)
```
