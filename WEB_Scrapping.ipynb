```python
# find the links in the website.
# find the anchor tags in the website.
# find the images in the websites.
# try to beautify the extract content.
# find paragraphs
# find titles 

def getHTML(url):
    response = requests.get(url)
    return response.text

scrape_url = 'https://www.flipkart.com/'
html_document = getHTML(scrape_url)
soup = BeautifulSoup(html_document,'html.parser')

# loop 
for link in soup.find_all('a'):
    print(link.get('href'))
```

```python
for tag in soup.find_all('a'):
    print(tag.text)
```

```python
for i in soup.find_all('img'):
    print(i['src'])
```
